```python
import json
```

```python
def find_single_quote_indices(input_string):
    # Initialize an empty list to store the indices of single quotes
    single_quote_indices = []

    # Iterate through the input string
    for index, char in enumerate(input_string):
        # Check if the character is a single quote
        if char == "'":
            # Add the index to the list
            single_quote_indices.append(index)
    return single_quote_indices
```

```python
def backslash(input_string):
    # Initialize an empty list to store the indices of single quotes
    backslash_indices = []

    # Iterate through the input string
    for index, char in enumerate(input_string):
        # Check if the character is a single quote
        if char == "\\":
            # Add the index to the list
            backslash_indices.append(index)
    # print(backslash_indices)
    return backslash_indices
```

```python
def replace_single_quotes_with_double(input_string):
    # Find the indices of single quotes
    indices_to_replace = find_single_quote_indices(input_string)

    # Convert the string to a list of characters
    text_list = list(input_string)

    # Replace single quotes with double quotes

    for index in indices_to_replace[:2]:
        text_list[index] = '"'
    
    # Replace single quotes with empty value
    for index in indices_to_replace[2:]:
        text_list[index] = ''
        
    # Join the list back into a string
    modified_text = ''.join(text_list)
    

    return modified_text
```

```python
def replace_backslash(input_string):
    # Find the indices of single quotes
    indices_to_replace = backslash(input_string)

    # Convert the string to a list of characters
    text_list = list(input_string)
    # print(text_list[624:638])
    
    for index in indices_to_replace:
        if index == index+1:
            text_list[index] = ''
            text_list[index+1] = ''
            text_list[index+2] = ''
        if index != index+1:
            text_list[index] = ''
            text_list[index+1] = ''
        
    # Join the list back into a string
    modified_text = ''.join(text_list)

    return modified_text
```

```python
def process_json_file(file_path):
    # Read the JSON file as a string
    with open(file_path, 'r') as file:
        json_string = file.read()

    modified_json_string = replace_backslash(json_string)
    modified_json_string = replace_single_quotes_with_double(modified_json_string)
    # print(modified_json_string)
    
    # Return the modified JSON string
    return modified_json_string
```

```python
def export_json_to_file(modified_json_string, file_path):
    # Convert the JSON string to a Python dictionary
    data_dict = json.loads(modified_json_string)

    # Write the JSON data to the file
    with open(file_path, 'w') as json_file:
        json.dump(data_dict, json_file)

    print("JSON data has been exported to", file_path)
```

```python
# List of numbers for the loop
numbers = list(range(0, 1000))

# List to store file names
# file_names = []

# Loop through each number and generate the file names
for number in numbers:
    file_name = r"C:\Zii\Notebook\Inixindo\DocData-json0\DocData_Doc{}.json".format(number)
    # file_names.append(file_name)

    # Call the function and pass the file path as an argument
    modified_json_string = process_json_file(file_name)
    
    # Define the file path to save the JSON data
    file_path = r'C:\Zii\Notebook\Inixindo\after\New_DocData_Doc{}.json'.format(number)

    # Call the function to export the JSON data to the file
    export_json_to_file(modified_json_string, file_path)
```
